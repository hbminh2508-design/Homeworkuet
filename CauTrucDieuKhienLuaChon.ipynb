W3A1: Viết một chương trình nhập vào một số nguyên n và in ra các chữ số đảo ngược nhau.

```python
n = int(input("Vui lòng nhập một số nguyên n:  "))
try:
  danh_sach = [int(i) for i in str(n)]
  danh_sach_dao = danh_sach[::-1] #Bỏ qua start | stop mà tiến qua step -> nhảy ngược thì số bị đảo ngược
  so_dao = int("".join(map(str, danh_sach_dao)))
  print(so_dao)
except ValueError:
  print("Vui lòng nhập vào một số nguyên!")
```

W3A3: Kiểm tra xem một số có phải là lũy thừa của 2 hay không.
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 hay không.
Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0

```python
#Ta nhận ra một cái bẫy là nếu n = 0 thì n&(n-1) == 0 nhưng 0 không phải luỹ thừa của 2
n = int(input("Vui lòng nhập một số nguyên n:  "))
def laLuyThuaCua2(n):
  return n > 0 and n & (n-1) == 0
laLuyThuaCua2(n)
```

W3A6: [Even and Odd]
Nhập vào một số nguyên dương x, bạn hãy kiểm tra xem x là số chẵn hay lẻ. Nếu x là số chẵn, in ra "Even". Nếu x
là số lẻ, in ra "Odd".


```python
x = int(input("Vui lòng nhập một số nguyên dương x: "))
def PhanBietChanLe(x):
  if x <= 0:
    return "Vui lòng nhập vào một số nguyên dương!"
  else:
    if x % 2 == 0:
      return "Even"
    else:
      x % 2 != 0
      return "Odd"

print(PhanBietChanLe(x))
```

W3A11: [Phân loại tam giác]
Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm tra 3 số đã nhập có tạo thành một tam giác hợp
lệ hay không. Nếu không in ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều", "Tam giác cân", hay
"Tam giác thường".

```python
sides = sorted(map(int, input("Nhập vào 3 số nguyên dương: ").split()))
a, b, c = sides
def KiemTraTamGiac(sides):
  if a + b > c:
    loai = len(set(sides))
    if loai == 1:
      return "Tam giác đều"
    elif loai == 2:
      return "Tam giác cân"

    return "Tam giác thường"

  return "Không phải tam giác"

print(KiemTraTamGiac(sides))
```

W3A12: [Kiểm tra năm nhuận]
Nhập vào một số nguyên dương n là số năm, kiểm tra xem năm đó có phải là năm nhuận hay không. Năm nhuận là năm
chia hết cho 4 nhưng không chia hết cho 100, hoặc chia hết cho 400. Nếu là năm nhuận in ra "Yes", nếu không in ra "No"

W3A13: [Tính tiền điện]
Nhập vào số kWh điện tiêu thụ. Tính tiền điện theo các bậc sau:
- 0-50 kWh: 1500đ/kWh
- 51-100 kWh: 2000đ/kWh
- 100 kWh: 3000đ/kWh


```python
soDien = int(input("Nhập số KWh điện tiêu thụ: "))
def TinhTienDien(soDien):
  if soDien <= 50:
    return soDien * 1500
  elif 51 <= soDien <= 100:
    return (50*1500) + (soDien - 50) * 2000
  else:
    return (50 * 1500) + (50 * 2000) + (soDien - 100) * 3000

print(TinhTienDien(soDien))
```

W3A14: [Giải phương trình bậc nhất]
Nhập vào hai số thực a và b của phương trình ax + b = 0. Tìm và in ra nghiệm của phương trình, kết quả làm tròn đến số
thập phân thứ 2. In ra "Vô nghiệm" hoặc "Vô số nghiệm" trong trường hợp tương ứng.


```python
a, b = map(int, input("Nhập vào a và b cách nhau bởi dấu cách: ").split())
def GiaiPhuongTrinhBacNhat(a, b):
  if a == 0:
    if b == 0:
      return "Vô số nghiệm"
    return "Vô nghiệm"

  else:
    phepChia = round(-b/a, 2)
    return phepChia

print(GiaiPhuongTrinhBacNhat(a, b))
```

W3A15: [Xếp loại học lực]
Nhập vào điểm trung bình của một học sinh và in ra học lực của học sinh đó. Xếp loại học lực theo quy tắc:
- >= 8.0: Giỏi
- >= 6.5: Khá
- >= 5.0: Trung bình
- < 5.0: Yếu


```python
n = float(input("Nhập vào điểm trung bình của học sinh: "))
match n:
  case n if n >= 8.0:
    print("Giỏi")
  case n if n >= 6.5:
    print("Khá")
  case n if n >= 5.0:
    print("Trung bình")
  case n if n < 5.0:
    print("Yếu")
```

W3A16: [Rounding]
Nhập vào một số thực: in ra kết quả làm tròn lên ,tròn xuống, làm tròn tới số nguyên gần nhất (không được sử dụng hàm
sẵn có của pyton). Các kết quả cách nhau một dấu cách


```python
n = float(input("Nhập vào một số thực: "))
val = n % 1 #VD: n = 3,14 thì 3,14 mod 1 = 0,14 còn n = -3,14 thì -3,14 mod 1 = 0.86 (Lấy phần dương bù lại)

#Làm tròn xuống
floor_val = n - val #VD: 3.14 - 0.14 = 3 | -3.14 - 0.86 = -4

#Làm tròn lên
if val == 0:
  ceil_val = floor_val
else:
  ceil_val = floor_val + 1

#làm tròn đến giá trị gần nhất
if val >= 0.5:
  round_val = ceil_val
else:
  round_val = floor_val

print(f"{int(ceil_val)} {int(floor_val)} {int(round_val)}")

```
