Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là độ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn hình.

```python
#Cách 1
import math

a, b, c = map(float, input("Nhập 3 cạnh tam giác: ").strip().split())

def kiemTraTamGiac(a, b, c):

    if a + b > c and a + c > b and b + c > a:
        print('La ba canh cua mot tam giac')
        p = (a + b + c) / 2

        if a == b == c:
            print('Tam giac deu')
        elif a == b or a == c or b == c:
            if (round(a**2 + b**2, 2) == round(c**2, 2) or
                round(a**2 + c**2, 2) == round(b**2, 2) or
                round(b**2 + c**2, 2) == round(a**2, 2)):
                print('Tam giac vuong can')
            else:
                print('Tam giac can')
        elif (round(a**2 + b**2, 2) == round(c**2, 2) or
              round(a**2 + c**2, 2) == round(b**2, 2) or
              round(b**2 + c**2, 2) == round(a**2, 2)):
            print('Tam giac vuong')
        else:
            print('Tam giac thuong')

        dienTich = math.sqrt(p * (p - a) * (p - b) * (p - c))
        chuVi = a + b + c
        print(f'Chu vi: {chuVi}')
        print(f'Dien tich: {dienTich}')
    else:
        print('Khong la ba canh cua mot tam giac')

kiemTraTamGiac(a, b, c)
```

```python
#Cách 2
import math

def giai_tam_giac():
    try:
        #sorted: Để sắp xếp các số từ bé đến lớn.
        sides = sorted(map(float, input("Nhập 3 cạnh: ").split()))
        a, b, c = sides

        if a + b <= c:
            return print("Khong la ba canh cua mot tam giac")

        loai = "thuong"
        '''
        Dùng để xem có bao nhiêu cạnh khác nhau từ đó phân loại được các tam giác.
        - set: Chứa các phần tử không trùng lặp
        - len: Đếm các phần tử còn lại trong tập
        '''
        count = len(set(sides))
        is_vuong = round(a**2 + b**2, 2) == round(c**2, 2)

        if count == 1: loai = "deu"
        elif count == 2: loai = "can-vuong" if is_vuong else "can"
        elif is_vuong: loai = "vuong"

        p = sum(sides) / 2
        s = math.sqrt(p * (p-a) * (p-b) * (p-c))

        print(f"La tam giac {loai}\nChu vi: {p*2}\nDien tich: {s}")
    except:
        print("Dữ liệu nhập không hợp lệ")

giai_tam_giac()
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
import math
def giaiPTBAC2():
  try:
    a, b, c = map(float, input("Nhập 3 số a, b và c (Cách nhau bởi các dấu cách): ").strip().split())
    if a == 0:
      if b == 0:
        print("Phương trình vô số nghiệm" if c == 0 else "Phương trình vô nghiệm")
      else:
        print(f"Phương trình có nghiệm duy nhất x = {-c/b}")
    else:
      delta = b**2 - 4*a*c
      if delta < 0:
        print("Phương trình vô nghiệm")
      elif delta == 0:
        print(f"Phương trình có nghiệm kép x1 = x2 = {-b/(2*a)}")
      else:
        x1 = (-b + math.sqrt(delta)) / (2*a)
        x2 = (-b - math.sqrt(delta)) / (2*a)
        print(f"Phương trình có 2 nghiệm phân biệt:\nx1 = {x1}\nx2 = {x2}")

  except ValueError:
    print("Dữ liệu nhập không hợp lệ")

giaiPTBAC2()
```

Bài 8: Giải hệ phương trình tuyến tính:
ax + by = m
cx + dy = n

```python
''' Thuật toán (Quy tắc Cramer)
3 định thức D, D_x, và D_y:
D = a.d - b.c
D_x = m.d - b.n
D_y = a.n - m.c

Nếu D != 0 thì hệ có nghiệm duy nhất x = D_x/D và y = D_y/D
Nếu D = 0:
  Nếu D_x = 0 và D_y = 0 thì hệ có vô số nghiệm (Hai đường thẳng trùng nhau)
  Nếu D_x = 0 hoặc D_y = 0 thì hệ vô nghiệm (Hai đường thẳng song song)
'''

try:
    a, b, m = map(float, input("Nhập vào phương trình 1 (a, b, m): ").split())
    c, d, n = map(float, input("Nhập vào phương trình 2 (c, d, n): ").split())
    print(f"Giải hệ phương trình:\n{a}x + {b}y = {m}\n{c}x + {d}y = {n}")

    D = a*d - b*c
    Dx = m*d - b*n
    Dy = a*n - m*c

    if D != 0:
        x = Dx/D
        y = Dy/D
        print(f"Hệ phương trình có nghiệm duy nhất: x = {x}, y = {y}")
    else:
        if Dx == 0 and Dy == 0:
            print("Hệ phương trình có vô số nghiệm")
        else:
            print("Hệ phương trình vô nghiệm")
except ValueError:
    print("Vui lòng nhập đầy đủ các số cách nhau bởi dấu cách.")
```

Bài 9: Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao nhiêu giây.

```python
#Cách 1
x = int(input("Nhập thời gian cho công việc này theo đơn vị giây: "))
gio = x//3600
phut = (x-(gio*3600))//60
giay = x - (gio*3600) - (phut*60)
print(f"Thời gian công việc này là: {gio} giờ | {phut} phút | {giay} giây.")
```

```python
#Cách 2: Hàm divmod(a, b) thực hiện cả phép chia lấy phần nguyên và chia lấy phần dư
x = int(input("Nhập số giây: "))

gio, giay_du = divmod(x, 3600)
phut, giay = divmod(giay_du, 60)

print(f"{gio} giờ | {phut} phút | {giay} giây")
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
import math as M
def square(x, y):
  return (y - x)**2
R = float(input("Nhập bán kính đường tròn vào đây: "))
Xo, Yo = map(float, input("Nhập toạ độ điểm O (Xo, Yo): ").split())
Xa, Ya = map(float, input("Nhập toạ độ điểm A (Xa, Ya): ").split())
d = M.sqrt(square(Xa, Xo) + square(Ya, Yo))
if d <= R:
  print("Điểm A thuộc đường tròn")
else:
  print("Điểm A không thuộc đường tròn")
```

```python
#Cách 2: DÙng hàm math.hypot(dx, dy)
import math

Xo, Yo = map(float, input("Nhập tọa độ tâm O (x y): ").split())
R = float(input("Nhập bán kính R: "))

Xa, Ya = map(float, input("Nhập tọa độ điểm A (x y): ").split())

d = math.hypot(Xa - Xo, Ya - Yo)

if d <= R:
    print(f"Khoảng cách d = {d:.2f} <= R. Điểm A thuộc đường tròn.")
else:
    print(f"Khoảng cách d = {d:.2f} > R. Điểm A không thuộc đường tròn.")
```
