Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím
Ví dụ: Ta nhập h = 4:
image.png

```python
def ve_tam_giac_can(h):
  for i in range(1, h + 1):
    print(" " * (h - i), end ="")

    if i == 1:
      print("*")

    elif i == h:
      print("*" * (2*i - 1))

    else:
      print("*" + " " * (2*i - 3) + "*")

#_______________________________________________________________________________
h = int(input("Vui lòng nhập độ cao h tam giác: "))
if h > 0:
  ve_tam_giac_can(h)
else:
  print("Vui lòng nhập độ cao hợp lệ")
```

Bài 6: Nhập vào số nguyên dương a, in ra tất cả các ước dương của a

```python
def in_uoc_so(a):
  if a <= 0:
    print("Số không hợp lệ")
    return

  uoc_so = []
  for i in range(1, a + 1):
    if a % i == 0:
      uoc_so.append(i)

  print(f"Các ước nguyên dương của số {a} là: ", end="")
  print(*uoc_so, sep = ", ")

a = int(input("Vui lòng nhập vào số nguyên dương: "))
in_uoc_so(a)
```

Bài 8: Nhập vào số nguyên dương a và b, in toàn bộ ước dương chung của a và b

```python
import math as toan
def in_ucln(a, b):
  if a <= 0 or b <= 0:
    print("Vui lòng nhập số hợp lệ !")
    return

  uoc_chung = set()
  ucln = toan.gcd(a, b)
  for i in range(1, toan.isqrt(ucln) + 1):
    if ucln % i == 0:
      uoc_chung.add(i)
      uoc_chung.add(ucln // i)

  print(f"Các ước chung dương của {a} và {b} là: ", end="")
  print(*uoc_chung, sep = ", ")

a = int(input("Vui lòng nhập vào số nguyên dương a: "))
b = int(input("Vui lòng nhập vào số nguyên dương b: "))
in_ucln(a, b)
```

Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không.
Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28…

```python
import math as toan
def so_hoan_hao(a):
  if a <= 0:
    print("Vui lòng nhập số hợp lệ !")
    return

  tong = 0
  for i in range(1, toan.isqrt(a) + 1):
    if a % i == 0:
      tong += i
      if i != a // i:
        tong += a // i

  if tong == a:
    print(f"{a} là số hoàn hảo")
  else:
    print(f"{a} không là số hoàn hảo")
```

Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
a = int(input("Nhập số nguyên dương A: "))

n = 0
tong_n = 0

while tong_n <= a:
  n += 1
  tong_n += 1/n

print(n)
```

Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng

```python
a = int(input("Nhập vào một số nguyên: "))

if a == -1:
  print("Dừng chương trình ngay lập tức")
else:
  max = a
  min = a

  while True:
    a = int(input("Nhập vào một số nguyên: "))

    if a == -1:
      break

    if a > max:
      max = a

    if a < min:
      min = a

  print(f"Số lớn nhất là {max} và số bé nhất là {min}.")
```

Bài 15: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số. Làm bằng nhiều
cách nhất có thể, ko dùng mảng.

```python
n = int(input("Nhập vào số nguyên dương n: "))

sodem = 0
if n == 0:
    sodem = 1
else:
  while n > 0:
    n //= 10
    sodem += 1

print(f"n có {sodem} chữ số.")
```

Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu
chữ số lẻ

```python
n = int(input("Nhập vào số nguyên dương n: "))

so_chan = 0
so_le = 0
temp = n

if temp == 0:
  so_chan = 1

else:
  while temp > 0:
    chu_so = temp % 10

    if chu_so % 2 == 0:
      so_chan += 1

    else:
      so_le += 1

    temp //= 10

print(f' Số lượng số chẵn là: {so_chan}')
print(f' Số lượng số lẻ là: {so_le}')
```
