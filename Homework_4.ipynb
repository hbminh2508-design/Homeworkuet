W4A18: Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b

```python
import math as toan
a = int(input("Nhập vào số nguyên dương a: "))
b = int(input("Nhập vào số nguyên dương b: "))

ucln = toan.gcd(a, b)
uoc_chung = [i for i in range(1, ucln + 1) if ucln % i == 0]

print(uoc_chung)
```

W4A19: [EvenNumBers]
Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để in ra các số
dương chẵn nhỏ hơn hoặc bằng n (n <= 1000)

```python
so = 0
while so <= 1000:
  if so % 2 == 0 and so > 0:
    print(so, end = " ")
  so += 1
```

W4A21: Nhập vào số nguyên dương n, tính tổng các chữ số của n

```python
n = int(input("Nhập vào số nguyên dương n: "))
tong = 0
while n > 0:
  tong += n % 10
  n //= 10
print(tong)
```

W4A22: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python

n = int(input("Nhập vào số nguyên dương n: "))

so_chan = 0
so_le = 0
temp = n

if temp == 0:
  so_chan = 1

else:
  while temp > 0:
    chu_so = temp % 10

    if chu_so % 2 == 0:
      so_chan += 1

    else:
      so_le += 1

    temp //= 10

print(f' Số lượng số chẵn là: {so_chan}')
print(f' Số lượng số lẻ là: {so_le}')
```

W4A23: Nhập n
Cho S(k) = 1 + 2 + 3 + … + k
Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n

```python
n = int(input("Nhập số nguyên dương n: "))
k = 0
tong_so_k = 0

while tong_so_k + (k + 1) < n:

  k += 1
  tong_so_k += k

print(k)
```

W4A24: Nhập vào A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
a = int(input("Nhập số nguyên dương A: "))

n = 0
tong_n = 0

while tong_n <= a:
  n += 1
  tong_n += 1/n

print(n)
```

W4A25: Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
a = int(input("Nhập vào một số nguyên: "))

if a == -1:
  print("Dừng chương trình ngay lập tức")
else:
  max = a
  min = a

  while True:
    a = int(input("Nhập vào một số nguyên: "))

    if a == -1:
      break

    if a > max:
      max = a

    if a < min:
      min = a

  print(f"Số lớn nhất là {max} và số bé nhất là {min}.")
```

W4A26: [Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A

```python
n = int(input("Nhập số nguyên dương A: "))

a = 1
b = 1
c = a + b

while c <= n:
  a = b
  b = c
  c = a + b

print(b)
```
