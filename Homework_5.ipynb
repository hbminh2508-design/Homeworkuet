xạ một kí tự nghĩa là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ tự của các kí tự
được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có
thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành
'z', 'b' thành 'b' và 'c' thành 'x'.
Viết hàm trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai
từ đầu vào có cùng độ dài.

```python
a, b = map(str, input().split())

def Is_Isomorphic(a, b):
  if len(a) != len(b):
    return False

  map_a_to_b = {}
  map_b_to_a = {}

  for char_a, char_b in zip(a, b):
    if char_a in map_a_to_b and map_a_to_b[char_a] != char_b:
      return False

    if char_b in map_b_to_a and map_b_to_a[char_b] != char_a:
     return False

    map_a_to_b[char_a] = char_b
    map_b_to_a[char_b] = char_a

  return True

print(Is_Isomorphic(a, b))
```

0 <= |r| < |divisor|.

```python
def divide_with_remainder(dividend, divisor):
  if divisor == 0:
    return "Không tính được!"

  #Tính thương số q
  q = dividend // divisor

  #Tính số dư r
  r = dividend % divisor

  return q, r

dividend, divisor = map(int, input().split())
print(divide_with_remainder(dividend, divisor))
```

W5A24: Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.

```python
def are_collinear(x1, y1, x2, y2, x3, y3):
  '''
  A(x1, y1), B(x2, y2), C(x3, y3)
  AB = (x2 - x1, y2 - y1)
  AC = (x3 - x1, y3 - y1)

  Thẳng hàng <=> AB.AC = 0 <=> (x2 - x1)(y3 - y1) - (x3 - x1)(y2 - y1) = 0
  '''

  vector_AB = (x2 - x1, y2 - y1)
  vector_AC = (x3 - x1, y3 - y1)

  if vector_AB[0] * vector_AC[1] - vector_AB[1] * vector_AC[0] == 0:
    return True

  return False

x1, y1, x2, y2, x3, y3 = map(int, input().split())
print(are_collinear(x1, y1, x2, y2, x3, y3))
```

W5A25: Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3),
(0,3)], p = (2,1)

```python
#Kiểm tra nằm bên trái hay bên phải
def check_left_or_right(A, B, P):
  '''
  Công thức tích có hương 2 vector:
  AB = (xb - xa, yb - ya)
  AP = (xp - xa, yp - ya)
  AB.AP = (xb - xa)(yp - ya) - (yb - ya)(xp - xa)

  Khi đó:
  > 0 => P nằm bên trái AB
  < 0 => P nằm bên phải AB
  = 0 => Nằm ngay trên AB
  '''

  return (B[0] - A[0]) * (P[1] - A[1]) - (B[1] - A[1]) * (P[0] - A[0])

def point_in_quadrilateral(quad, p):

  # quad[0], quad[1], quad[2], quad[3] lần lượt là 4 đỉnh A, B, C, D
  d1 = check_left_or_right(quad[0], quad[1], p)
  d2 = check_left_or_right(quad[1], quad[2], p)
  d3 = check_left_or_right(quad[2], quad[3], p)
  d4 = check_left_or_right(quad[3], quad[0], p)

  co_am = (d1 < 0) or (d2 < 0) or (d3 < 0) or (d4 < 0)
  co_duong = (d1 > 0) or (d2 > 0) or (d3 > 0) or (d4 > 0)

  return not(co_am and co_duong)

print("===NHẬP TOẠ ĐỘ 4 ĐỈNH===")
quad = []
for i in range(4):
  x, y = map(int, input(f"NHẬP ĐỈNH {i+1}: ").split())
  quad.append((x, y))

print("===NHẬP TOẠ ĐỘ ĐIỂM MUỐN KIỂM TRA===")
p = []
x, y = map(int, input().split())
p = (x, y)

print(point_in_quadrilateral(quad, p))
```

W5A28: Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).

```python
import math as toan

def Calculate_Angle(u, v):
  #Tính tích vô hướng
  tich_vo_huong = u[0] * v[0] + u[1] * v[1]

  #Tính độ dài u và v
  do_dai_u = toan.sqrt(u[0] ** 2 + u[1] ** 2)
  do_dai_v = toan.sqrt(v[0] ** 2 + v[1] ** 2)

  #Tính góc
  goc_radian = toan.acos(tich_vo_huong / (do_dai_u * do_dai_v))
  goc_degree = toan.degrees(goc_radian)

  return goc_degree

print("===NHẬP TOẠ ĐỘ 3 ĐỈNH===")
ux, uy = map(int, input("Nhập tọa độ điểm U: ").split())
vx, vy = map(int, input("Nhập tọa độ điểm V: ").split())

U = (ux, uy)
V = (vx, vy)

print(round(Calculate_Angle(U, V), 2))
```

W5A30: Viết hàm tính hệ số nhị thức C(n, k).

```python
# C(n, k) = n! / k!(n-k)! = (n*(n-1)*...*(n-k+1))/(k*(k-1)*...*1)

n, k = map(int, input().split())

def nCr_Calculate(n, k):
  if k > n - k:
    k = n - k

  ketqua = 1
  for i in range(1, k + 1):
    ketqua *= (n - i + 1) / i

  return int(ketqua)

print(nCr_Calculate(n, k))
```

W5A34: Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau
Nếu n ≤ 0 trả về 0.

```python
def sum_1_to_n(n):

  if n <= 0:
      return 0

  tong = 0
  for x in range(1, n + 1):
      chuoi_x = str(x)
      if len(set(chuoi_x)) == len(chuoi_x):
          tong += x

  return tong

n = int(input())
print(sum_1_to_n(n))
```
