W6A1: Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng
giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.
Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]

```python
danh_sach = list(map(int, input().split()))
ket_qua = []
for x in danh_sach:
  if x not in ket_qua:
    ket_qua.append(x)
print(ket_qua)
```

W6A3:Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple
sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp
xếp.)

```python
t = tuple(map(int, input().split()))
k = int(input())

n = len(t)
if n > 0:
  k = k % n
  results = t[k:] + t[:k]
else:
  results = t

print(results)
```

W6A4: Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key
lặp lại, thêm vào list của key đó


```python
data_list = input().split()

mapping = {}

for pair in data_list:

  key, value = pair.split(':')

  if key in mapping:
    mapping[key].append(value)
  else:
    mapping[key] = [value]

print(mapping)
```

W6A5: Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ...,
'zeros': ...} (đếm số >0, <0, =0)

```python
data_list = list(map(int, input().split()))

count_positives = 0
count_negatives = 0
count_zeros = 0

for num in data_list:
  if num > 0:
    count_positives += 1
  elif num < 0:
    count_negatives += 1
  else:
    count_zeros += 1

results = {
    'positives':count_positives,
    'negatives':count_negatives,
    'zeros':count_zeros
}
print(results)
```

W6A9: Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một
dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ
được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới
theo thứ tự từ điển tăng dần của key.

```python
def get_dict():
  data_list = input().split()
  d = {}
  for pair in data_list:
    if ':' in pair:
      key, value = pair.split(':')
      d[key] = int(value)
  return d

print("Nhập vào dict1 (key:value): ")
dict1 = get_dict()
print("Nhập vào dict2 (key:value): ")
dict2 = get_dict()

results = dict1.copy() #Copy dic1 vào kqua

for key, value in dict2.items():
  #So sánh các key của 2 dict
  if key in results:
    results[key] += value
  else:
    results[key] = value

sorted_key = sorted(results)
f_results = {k: results[k] for k in sorted_key}

print(f_results)
```

W6A10: Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên
dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là
một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh
sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được
sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.

```python
numbers = list(map(int, input().split()))
k = int(input())

result = []

n = len(numbers)
for i in range(n):
  for j in range(i+1, n):
    if numbers[i] + numbers[j] == k:
      cap_so = (numbers[i], numbers[j])

      if cap_so not in result:
        result.append(cap_so)

result.sort()

print(result)
```

W6A12: [Tìm phần tử xuất hiện nhiều nhất]
Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong
list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ
nhất trong số đó.

```python
danh_sach = list(map(int, input().split()))

count = {}
for x in danh_sach:
  if x in count:
    count[x] += 1
  else:
    count[x] = 1

max_freq = 0
result = None

for num, freq in count.items():
  if freq > max_freq:
    max_freq = freq
    result = num

  elif freq == max_freq:
    if num < result:
      result = num

print(result)
```

W6A13: [Đảo ngược key và value của dictionary]
Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo
một dictionary mới bằng cách đảo ngược key và value của dictionary ban
đầu. In ra dictionary mới.

```python
def get_dict():
  data_list = input().split()
  d = {}
  for pair in data_list:
    if ':' in pair:
      key, value = pair.split(':')
      d[key] = int(value)

  return d

dict1 = get_dict()

dict2 = {value: key for key, value in dict1.items()}

print(dict2)
```

W6A14: [Giao của hai danh sách]
Nhập vào hai dòng, mỗi dòng là một danh sách các số nguyên. Hãy tìm các
phần tử chung (giao) của hai danh sách này và in ra một list mới chứa các
phần tử chung đó, không trùng lặp.

```python
n1 = list(map(int, input().split()))
n2 = list(map(int, input().split()))
n = []

for i in range(len(n1)):
  for j in range(len(n2)):
    if n1[i] == n2[j] and n1[i] not in n:
      n.append(n1[i])

print(n)
```

W6A15: [Lọc dictionary theo giá trị]
Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số
nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp
key-value từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary
mới.

```python
str_dict = input()
my_dict = {}
part_dict = str_dict.split()

k = int(input())

for i in range(0, len(part_dict), 2):
  key = part_dict[i]
  value = int(part_dict[i + 1])
  my_dict[key] = value


results = {key: value for key, value in my_dict.items() if value > k}

print(results)
```
