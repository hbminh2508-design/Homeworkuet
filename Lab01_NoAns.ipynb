### Problem 8
#### Write a function that returns the character that is n positions after character x in the alphabet.
#### Preserve case and wrap around within A..Z or a..z.

```python
x = str(input())
n = int(input())

def dichChuyen(x, n):
  if 0 <= n <= 25:
    acii = ord(x)

    #Thao tác với chữ thường:
    if 65 <= acii <= 90:
      return chr((acii - 65 + n)%26 + 65)
    #Thao tác với chữ hoa:
    elif 97 <= acii <= 122:
      return chr((acii - 97 + n)%26 + 97)

print(dichChuyen(x, n))
```

### Problem 9
#### Write a function that converts an uppercase letter to lowercase. If the input is not a single character, raise ValueError.

```python
c = str(input())
def vietHoaSangVietThuong(c):
  if len(c) != 1:
    raise ValueError
  else:
    chuThuong = c.lower()
    return chuThuong

print(vietHoaSangVietThuong(c))
```

### Problem 10
#### Write a function that, given a lowercase letter, returns a string containing that letter repeated 10 times.

```python
c = str(input())
def traVeGapMuoi(c):
  if c.islower:
    nhanLen = 10 * c
  else:
    return ValueError
  return nhanLen

print(traVeGapMuoi(c))
```

### Problem 12
#### Write a function that returns a right isosceles triangle of height n using '*'.
#### Line i (1-indexed) has i '*', joined with newlines.
#### Example with n = 4
#### *
#### **
#### ***
#### ****

```python
n = int(input())
def chenDauSao(n):
  if n <= 0:
    return ""
  else:
    lines = [("*"*i) for i in range (1, n+1)]
  return "\n".join(lines)
print(chenDauSao(n))
```
