W4A1: [SumOfNumBers]
Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để tính tống các số nguyên từ 1 đến n. (n <= 1000)


```python
n = int(input("Vui lòng nhập vào số nguyên n để tính tổng từ 1 đến n: "))
sodem = 0
tong = 0
if n > 1000:
  print("Vui lòng nhập lại số n <= 1000")
else:
  while sodem < n:
    sodem += 1
    tong = sodem + tong
  print(tong)
```

```python
while True:
    n = int(input("Vui lòng nhập vào một số nguyên: "))

    if n < 0:
        print("Số âm không xét tính nguyên tố. Vui lòng nhập lại!")
    else:
        if n < 2:
            print(f"{n} không phải là số nguyên tố")
        elif n == 2:
            print("2 là số nguyên tố chẵn duy nhất")
        else:
            for i in range(2, n):
                if n % i == 0:
                    print(f"{n} không phải là số nguyên tố (vì chia hết cho {i})")
                    break
            else:
                print(f"{n} là số nguyên tố")

        break
```

W4A6: "[SumPrimesInRange]
Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn
[a, b]. Mỗi dòng là một bộ test (a b)."

```python
def isPrime(n):
  if n < 2:
    return False

  for i in range(2, int(n ** 0.5) + 1):
    if n % i == 0:
      return False

    return True

a, b = sorted(map(int, input().split()))
sum_prime = 0
for k in range (a, b + 1):
  if isPrime(k):
    sum_prime += k

print(sum_prime)
```

W4A7: "[LargestPrimeFactor]
Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n."

```python
n = int(input("Vui lòng nhập số nguyên dương n lớn hơn 2: "))

while n % 2 == 0:
  max_prime = 2
  n //= 2

for i in range (3, int(n ** 0.5) + 1, 2):
  while n % i == 0:
    max_prime = i
    n //= i

if n > 1:
  max_prime = n

print(max_prime)
```

W4A8: [ReverseAddPalindrome]
Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được.

```python
#Palindrome: Một số đọc từ trái hay phải sang đều mang cùng một kết quả (VD: 121, 9999, 1881, 2002 ...)

def reverse(n):
  so_moi = 0
  while n > 0:
    so_moi = so_moi * 10 + n % 10
    n //= 10
  return so_moi

n = int(input())
so_buoc = 0

while True:
  reverse_n = reverse(n)

  if n == reverse_n:
    break

  n = n + reverse(n)
  so_buoc += 1

print(so_buoc, n)
```

W4A9: [PerfectSquaresDistinctDigits]
Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số
đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu
cách). Nếu không có, in ra (no number).

```python
n = int(input("Nhập vào một số nguyên dương n: "))
danh_sach = []
for i in range(0, n + 1):
  if i <= n and i ** 0.5 == int(i ** 0.5) and len(set(str(i))) == len(str(i)):
    danh_sach.append(i)

if not danh_sach:
  print("No number")
else:
  print(*danh_sach)
```

W4A10: "[LongestCollatzUpToN]
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất."

image.png

```python
n = int(input("Nhập vào số nguyên dương n: "))
max_l = 0
best_x = 0

for i in range (1, n+1):
  temp = i
  current_l = 1

  while temp > 1:
    if temp % 2 == 0:
      temp //= 2
    else:
      temp = 3 * temp + 1
    current_l += 1

  if current_l > max_l:
    max_l = current_l
    best_x = i

print(best_x, max_l)
```

W4A11: [EvenDivisors]
Viết chương trình đếm số lượng ước số chẵn của số nguyên dương n với n < 10^6.

```python
n = int(input("Nhập vào một số n: "))

sodem = 0

for i in range(1, int((n ** 0.5)) + 1):
  if n % i == 0:
    if i % 2 == 0:
      sodem += 1

    if (n // i) % 2 == 0 and i != n // i:
      sodem += 1

print(sodem)
```

W4A13: [ClosePairNumber]
Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật:
Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại.
Ví dụ 220 và 284 là một cặp số thân thiết
220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55,
110. Tổng của 11 ước này cho chúng ta 284
284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng
vừa đúng bằng 220
Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không.
Nếu có, in ra màn hình "true", ngược lại, in ra "false"

```python
a, b = map(int, input("Nhập vào hai số nguyên dương a và b: ").split())

danh_sach_a = []
danh_sach_b = []

for i in range(1, a):
  if a % i == 0:
    danh_sach_a.append(i)

for i in range(1, b):
  if b % i == 0:
    danh_sach_b.append(i)

if sum(danh_sach_a) == b and sum(danh_sach_b) == a:
  print("true")
else:
  print("false")
```

W4A14: [GCD]
Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình
tìm ước chung lớn nhất của hai số m và n.

```python
a, b = map(int, input("Vui lòng nhập vào 2 số a và b: ").split())

while b != 0:
  so_du = a % b
  a = b
  b = so_du

print(a)
```

"Vừa gà vừa chó,

Từ bài toán dân gian trên, mở rộng thành chương trình nhận đầu vào là tổng số
con và tổng số chân của gà và chó. Nếu tìm được số phù hợp, in ra số lượng gà và
số lượng chó. Ngược lại, in ra "invalid".

```python
tong_con = int(input("Nhập vào tổng số con: "))
tong_chan = int(input("Nhập vào tổng số chân: "))

tim_thay = False

for g in range(tong_con + 1):
  c = tong_con - g

  if tong_chan == (2 * g + 4 * c):
    print(f"Tổng số gà là {g} ")
    print(f"Tổng số chó là {c}")
    tim_thay = True
    break

if not tim_thay:
  print("invalid")
```
